```python
data = pd.read_csv('D:/Master ML/Data Mining/3- MPL/data/arcade-revenue-vs-cs-doctorates.csv')
data.head()

rev = data['Total Arcade Revenue (billions)']
cs = data['Computer Science Doctorates Awarded (US)']

data['Revenue per Person'] = cs + rev

z = rev /  cs 
data['Revenue per Person'] = z  * 1000000
per = data['Revenue per Person']

plt.figure()
plt.scatter(cs, per); 

plt.title('Damped exponential decay')
plt.xlabel('Computer Science Doctorates Awarded (US)')
plt.ylabel('Revenue per Person (billions)')
plt.subplots_adjust(left=0.15)
plt.show()
```
